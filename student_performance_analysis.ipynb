```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data import
df = pd.read_csv("student-mat.csv",sep=';')
```

```python
#Data sample
df.head()
```

Explore and Clean Data

```python
# Inspect dataset
print(df.shape)
print(df.dtypes)
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Remove duplicates
print("Duplicates before:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicates after:", df.duplicated().sum())
```

Analysis

```python
#Average Final Grade
print("Average Final Grade:", df['G3'].mean())

#Students Scoring Above 15
count = (df['G3'] > 15).sum()
print("students scoring above 15:", count)

#correlation Study Time vs Performance
correlation = df['studytime'].corr(df['G3'])
print("correlation between study time and final grade:", correlation)

#Gender-wise Performance
avg_by_gender = df.groupby('sex')['G3'].mean()
print(avg_by_gender)
```

Data Visualization

```python
# Histogram
plt.figure(figsize=(6,4))
sns.histplot(df['G3'], kde=True, color='skyblue')
sns.kdeplot(df['G3'], color='red')
plt.title("Distribution of Final Grades", color='purple')
plt.xlabel("Final Grade (G3)", color='green')
plt.ylabel("Count", color='blue')
plt.show()
```

```python
# Scatter plot: study time vs grade
plt.figure()
sns.scatterplot(x='studytime', y='G3', data=df, color='orange')
plt.title("Study Time vs Final Grade", color='blue')
plt.xlabel("Study Time", color='red')
plt.ylabel("Final Grade (G3)", color='green')
plt.show()
```

```python
avg_score = df.groupby('sex')['G3'].mean()

plt.figure()
avg_score.plot(kind = 'bar',color = ['skyblue','pink'])
plt.title("Average Score by Gender",color = 'green')
plt.xlabel("Gender",color = 'purple')
plt.ylabel("Average Final Grade(G3)",color = 'blue')
plt.show()   
```
